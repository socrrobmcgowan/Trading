```python
def get_data(symbol):
    r = requests.get(f'http://quotes.wsj.com/{symbol}/historical-prices/download?&num_rows=6299.041666666667&startDate={month}/{day}/{last_year}&endDate={month}/{day}/{year}')
    pr = r.text.split('\n')
    pr = [x.split(',') for x in pr]
    new_r = requests.get(f'https://api.iextrading.com/1.0/tops?symbols={symbol}')
    new_pr = json.loads(new_r.content)
    try:
        price = round(float(new_pr[0]['lastSalePrice']),2)
        new_data = [today_date,'',float(),float(),price,float(0)]
        pr[0] = new_data
    except:
        pr.pop(0)
    df = pd.DataFrame(pr,columns=['date','open','high','low','close','volume'])
    df.sort_index(ascending=False,inplace=True)
    df.set_index('date', inplace = True)
    return df
```
