```python
import numpy as np
import pandas as pd
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import csv
%matplotlib inline
```

```python
symbols = []
csv_read = csv.reader(open('companylist.csv'))
for row in csv_read:
    if row[0] != 'Symbol':
        symbols.append(row[0])
```

```python
master_df_list = []
for symbol in symbols:
    df = web.RobinhoodHistoricalReader(symbol)
    master_df_list.append(df)
```

```python
x = 1
print(len(master_df_list))
for data in master_df_list:
    try:
        data = data.read()
        df = pd.DataFrame(data)
        df = df.reset_index(level = 0, drop = True)
        df = df.reset_index(level = 0, drop = True)
        df.dropna(inplace = True)
        df['close_price'] = df['close_price'].astype(float)
        df['prev_price'] = df['close_price'].shift(1)
        df['change'] = df['close_price'] - df['prev_price']
        df['gain'] = df['close_price'] - df['prev_price']
        df['loss'] = df['close_price'] - df['prev_price']
        for i in df.index: 
            if df['change'][i] > 0 :
                df['gain'][i] = df['change'][i]
                df['loss'][i] = float(0.0)
            elif df['change'][i] < 0 : 
                df['gain'][i] = float(0.0)
                df['loss'][i] = abs(df['change'][i])
            else: 
                df['gain'][i] = float(0.0)
                df['loss'][i] = float(0.0)
        df['avg_gain'] = df['gain'].rolling(14).mean()
        df['avg_loss'] = df['loss'].rolling(14).mean()
        df['rs'] = df['avg_gain'] / df['avg_loss']
        df['rsi'] = 100 - (100/(1.0+df['rs']))
    except KeyError:
        next
    print(x)
    x+=1
```
